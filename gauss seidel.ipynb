```python
def seidel(a, x ,b):
    n = len(a)              
    for j in range(n):        
        d = b[j]                  
        for i in range(n):     
            if(j != i):
                d-=a[j][i] * x[i]
        x[j] = d / a[j][j]
    return x    
   
```

```python
#saisie les données
n = int(input("Donner nombre des variables"))
epsilon = float(input("Donner l'epsilon"))
max_iteration=int(input("donner max nombre d'itération"))
a = []                            
b = []
x = []   
test=True
for i in range(n):
    l=[]
    for j in range(n):
        l.append(int(input("Donner A{}{}".format(i+1,j+1))))
    a.append(l)                            
for i in range(n):
    b.append(int(input("Donner b{}".format(i+1))))
    x.append(0)
```

```python
import copy
x1=copy.copy(x)
while test and max_iteration>0:
    x = seidel(a, x, b)
    #on va afficher le résultat pour chaque itération
    print(x)
    i=0
    while(test and i<n):
        if(abs(x1[i]-x[i])<=epsilon):
            test=False
            break
        i=i+1
    x1=copy.copy(x)
    max_iteration-=1
```
